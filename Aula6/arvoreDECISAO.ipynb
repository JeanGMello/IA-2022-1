```python
import pandas as pd
pd.options.mode.chained_assignment = None

from sklearn.model_selection import train_test_split

linkArquivo = "https://archive.ics.uci.edu/ml/datasets/Contraceptive+Method+Choice"
```

```python
cols = ["WifeAge", "WifeEducation", "HusbandEducation", "NumberOfChildrenEverBorn", "WifeReligion", "WifeNowWorking","HusbandOccupation", "StandardOfLiving", "MediaExposure",  "ContraceptiveMethodUsed"]

df = pd.read_csv('cmc.data', names=cols, dtype=str)
df. dropna(inplace=True)
df.head(10)
```

```python
y = df['ContraceptiveMethodUsed']
y.head()
```

```python
df.drop(columns=['ContraceptiveMethodUsed'], inplace=True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3, random_state=0, stratify=y)
```

```python
y = y_train
y_str = y.astype('category')
```

```python
X_train.WifeAge = X_train.WifeAge.astype(int)
X_train.WifeEducation = X_train.WifeEducation.astype(int)
X_train.HusbandEducation = X_train.HusbandEducation.astype(int)
X_train.NumberOfChildrenEverBorn = X_train.NumberOfChildrenEverBorn.astype(int)
X_train.WifeReligion = X_train.WifeReligion.astype(int)
X_train.WifeNowWorking = X_train.WifeNowWorking.astype(int)
X_train.HusbandOccupation = X_train.HusbandOccupation.astype(int)
X_train.StandardOfLiving = X_train.StandardOfLiving.astype(int)
X_train.MediaExposure = X_train.MediaExposure.astype(int)

X_test.WifeAge = X_train.WifeAge.astype(int)
X_test.WifeEducation = X_train.WifeEducation.astype(int)
X_test.HusbandEducation = X_train.HusbandEducation.astype(int)
X_test.NumberOfChildrenEverBorn = X_train.NumberOfChildrenEverBorn.astype(int)
X_test.WifeReligion = X_train.WifeReligion.astype(int)
X_test.WifeNowWorking = X_train.WifeNowWorking.astype(int)
X_test.HusbandOccupation = X_train.HusbandOccupation.astype(int)
X_test.StandardOfLiving = X_train.StandardOfLiving.astype(int)
X_test.MediaExposure = X_train.MediaExposure.astype(int)

for c in X_train.columns:
    if type(X_train[c].iloc[0]) == str:
        X_train[c] = X_train[c].astype('category')
        X_train[c] = X_train[c].cat.codes
```

```python
for c in X_test.columns:
    # print(X_test[c].iloc[0])
    if type(X_test[c].iloc[0]) == str:
        X_test[c] = X_test[c].astype('category')
        X_test[c] = X_test[c].cat.codes
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0, max_depth=4)
clf.fit(X_train, y_train)
```
